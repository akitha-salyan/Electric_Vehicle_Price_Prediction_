```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Cheapestelectriccars-EVDatabase.csv")
df.head()
```

# Data Processing

```python
print(df.columns.tolist())
```

```python
#There are a lot of missing values in the data, so now let us fill the missing data with zero
#filling empty data with 0
df=df.fillna('0')
```

```python
#let us get the name of the car manufacturer from the car name.
# Extract manufacturer (first word from 'Name')
df['Manufacturer'] = df['Name'].str.split(' ', n=1, expand=True)[0]
df['Manufacturer']
```

```python
#If you want both manufacturer and model separately,  can do:
df[['Manufacturer', 'Model']] = df['Name'].str.split(' ', n=1, expand=True)
df[['Manufacturer', 'Model']]
```

```python
#above explaination
#| Name                             | Manufacturer | Model                       |
#| -------------------------------- | ------------ | --------------------------- |
#| Opel Ampera-e                    | Opel         | Ampera-e                    |
#| Renault Kangoo Maxi ZE 33        | Renault      | Kangoo Maxi ZE 33           |
#| Nissan Leaf                      | Nissan       | Leaf                        |
#| Audi e-tron Sportback 55 quattro | Audi         | e-tron Sportback 55 quattro |
#| Porsche Taycan Turbo S           | Porsche      | Taycan Turbo S              |
```

```python
#cleaning the data
#removing currency sign
# Remove currency symbols and commas, then convert to numeric
df['PriceinUK'] = (
    df['PriceinUK']
    .astype(str)                # convert everything to string first
    .str.replace('£', '', regex=False)
    .str.replace(',', '', regex=False)
    .replace('nan', None)       # handle NaN strings
)

# Convert to numbers (will keep NaN where not available)
df['PriceinUK'] = pd.to_numeric(df['PriceinUK'], errors='coerce')

# Print cleaned column
print(df['PriceinUK'])


```

```python
# Cleaning Germany prices using pandas (faster, shorter)
df['PriceinGermany'] = (
    df['PriceinGermany']
    .astype(str)                            # make sure all values are strings
    .str.replace('€', '', regex=False)      # remove €
    .str.replace(',', '', regex=False)      # remove commas
    .replace('nan', None)                   # replace 'nan' string with None
)

# Convert to numeric (floats, NaN stays as NaN)
df['PriceinGermany'] = pd.to_numeric(df['PriceinGermany'], errors='coerce')

# Print cleaned column
print("Cleaned Germany Prices (Vectorized Method):")
print(df['PriceinGermany'])
```

```python
#Data is cleaned.
df.head(15)
```

```python
#we will prepare the data for Fast Charging Speed.
# Clean Fast Charging Speed column directly
df['FastChargeSpeed'] = (
    df['FastChargeSpeed']
    .astype(str)                              # convert all to string
    .str.replace(' km/h', '', regex=False)    # remove " km/h"
    .str.replace('-', '0', regex=False)       # replace '-' with 0
)

df['FastChargeSpeed'] = pd.to_numeric(df['FastChargeSpeed'], errors='coerce')

# Print cleaned column
print("Cleaned Fast Charging Speeds (Vectorized):")
print(df['FastChargeSpeed'])
```

```python
#we proceed to prepare the data for efficiency.
# Clean Efficiency column directly with pandas
df['Efficiency'] = (
    df['Efficiency']
    .astype(str)                              # convert everything to string
    .str.replace(' Wh/km', '', regex=False)   # remove " Wh/km"
    .replace('nan', None)                     # handle missing
)

df['Efficiency'] = pd.to_numeric(df['Efficiency'], errors='coerce')

# Print cleaned column
print("Cleaned Efficiency values (Vectorized):")
print(df['Efficiency'])
```

```python
#we can do the data cleaning and preparation for other parameters.
# Clean Range (remove " km")
df['Range'] = (
    df['Range']
    .astype(str)
    .str.replace(' km', '', regex=False)
)
df['Range'] = pd.to_numeric(df['Range'], errors='coerce')

# Clean TopSpeed (remove " km/h")
df['TopSpeed'] = (
    df['TopSpeed']
    .astype(str)
    .str.replace(' km/h', '', regex=False)
)
df['TopSpeed'] = pd.to_numeric(df['TopSpeed'], errors='coerce')

# Clean Acceleration (remove " sec")
df['Acceleration'] = (
    df['Acceleration']
    .astype(str)
    .str.replace(' sec', '', regex=False)
)
df['Acceleration'] = pd.to_numeric(df['Acceleration'], errors='coerce')

# Clean Subtitle (extract kWh from text)
df['Subtitle'] = (
    df['Subtitle']
    .astype(str)
    .str.replace('Battery Electric Vehicle | ', '', regex=False)
    .str.replace(' kWh', '', regex=False)
)
df['Subtitle'] = pd.to_numeric(df['Subtitle'], errors='coerce')

#  Print cleaned columns
print("Cleaned Data:")
print(df[['Range','TopSpeed','Acceleration','Subtitle']])
```

```python
#renaming that column to KWH (because you already cleaned it and it now only contains the battery size in kWh).
df = df.rename(columns = {'Subtitle': 'KWH'})
```

```python
#the data preparation is done. Now,  can focus on having a look at the data and data analysis.
df.head()
```

# Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Select features
X = df[['KWH', 'Acceleration', 'TopSpeed', 'Range', 'Efficiency', 'FastChargeSpeed', 'PriceinGermany']]
#You select several features (columns) from your dataset df which you want to use for clustering.
#These features describe various attributes of electric vehicles, e.g., battery size (KWH), acceleration, range, etc.
```

```python
# Scale features before clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Why scale?
#Different features have different units and ranges (e.g., KWH might be 20–100, acceleration 3–15 seconds).
#Scaling standardizes the data to have mean = 0 and standard deviation = 1 for each feature.
#This prevents features with larger ranges from dominating the clustering process.
```

```python
# Fit KMeans
km = KMeans(n_clusters=5, random_state=42)
y = km.fit_predict(X_scaled)

#KMeans is a clustering algorithm that groups data into 5 clusters (n_clusters=5).
#fit_predict does two things:
#It fits the model to your scaled data (finds clusters).
#It assigns each data point to a cluster (labelled from 0 to 4).
#y stores cluster labels for each data point.
```

```python
# Plot clusters using first two features (KWH and Acceleration)
plt.figure(figsize=(10, 6))

plt.scatter(X_scaled[y == 0, 0], X_scaled[y == 0, 1], color='red', label='Cluster 1')
plt.scatter(X_scaled[y == 1, 0], X_scaled[y == 1, 1], color='green', label='Cluster 2')
plt.scatter(X_scaled[y == 2, 0], X_scaled[y == 2, 1], color='blue', label='Cluster 3')
plt.scatter(X_scaled[y == 3, 0], X_scaled[y == 3, 1], color='cyan', label='Cluster 4')
plt.scatter(X_scaled[y == 4, 0], X_scaled[y == 4, 1], color='purple', label='Cluster 5')

# plot a scatter plot using the first two features after scaling:
#Feature 0 = KWH
#Feature 1 = Acceleration
#For each cluster (y == 0 means all points assigned to cluster 0), you plot points in different colors.
#Each cluster gets a different color and label for clarity.



# Plot cluster centers
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='yellow', s=300, marker='X', label='Centroids')
#km.cluster_centers_ contains the coordinates of each cluster’s 
#centroid (center point) in the scaled feature space.


plt.xlabel('Battery Capacity (KWH) [scaled]')
plt.ylabel('Acceleration (0-100 km/h) [scaled]')
plt.title('KMeans Clustering of EVs')
plt.legend()
plt.show()
```

# Machine Learning Model For Price Prediction of Electric Vehicles:   

```python
######################   Machine Learning Model For Price Prediction of Electric Vehicles:    #####################

#We will predict the price of the vehicle based on all the parameters and data, one thing to be pointed is that many data points are missing. 
#In, the case of vehicle prices, many vehicles have prices just in Pounds, and some have prices just in euros. So, for the sake of simplicity,
#we take only the prices in UK Pound.
```

```python
#Considering 1 Euro = 0.85 Pound
#Let us see the number of empty values in Pounds.
#c=0
#for i in range(0, len(df["PriceinUK"])):
#    if (df["PriceinUK"][i]==0):
#        c+=1
#print('Empty Values:',c)

print('Empty values',df["PriceinUK"].eq(0).sum())
```

```python
#As we will modify some data, we put this setting.
pd.options.mode.chained_assignment = None
#It’s just telling pandas “don’t warn me when I change a slice of a DataFrame”.
```

```python
#You’re looping row by row and filling PriceinUK with converted German price (×0.85) whenever it’s 0.
#for i in range(0, len(df["PriceinUK"])):
#    if (df["PriceinUK"][i]==0):
#        val=df["PriceinGermany"][i]*0.85
#        df["PriceinUK"][i]=val

#simple way
df.loc[df["PriceinUK"] == 0, "PriceinUK"] = df["PriceinGermany"] * 0.85
#What this does:
#df1["PriceinUK"] == 0 → finds all rows where UK price is 0.
#.loc[rows, "PriceinUK"] → selects only those rows in PriceinUK.
#= df1["PriceinGermany"] * 0.85 → assigns the converted values directly.

#ref:You filled the zero UK prices with an estimated value from German prices, even though it’s not a perfect method. 
#It’s just a practical choice because you don’t have the real data.

```

```python
#!pip install category_encoders
```

```python
import category_encoders as ce
train_df=df
# create object of Ordinalencoding
encoder= ce.OrdinalEncoder(cols=['Drive'],return_df=True,
                           mapping=[{'col':'Drive',
'mapping':{'Front Wheel Drive':1,'Rear Wheel Drive':2,'All Wheel Drive':3}}])
#fit and transform train data 
df_train = encoder.fit_transform(train_df)#Original data

df_train.info()
```

```python
#Except for vehicle name and manufacturer name, all else is a numeric value, so no problem in computing.
```

```python
#Now, we take the input data, which is taken as X.
x = df_train.drop(['Name','PriceinGermany','PriceinUK','Manufacturer','Model'], axis=1)
x=x.values
```

```python
x
#we can see that it is a multidimensional array.
```

```python
#So, now we have to take in the target, that is “y”.
y=df_train['PriceinUK'].values
```

```python
#loking for dimensions
x.shape 
```

```python
y.shape

#we can see that both have 180 data points.
```

```python
#Now, proceeding to the train, test split.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=7)
#We kept the test set at 1/3rd of the data and 2/3rd of the data is train set.
```

```python
#create a Random Forest Regressor model.
from sklearn.ensemble import RandomForestRegressor
# create regressor object
model = RandomForestRegressor(n_estimators = 300, random_state = 0)
# fit the regressor with x and y data
model.fit(X_train, y_train)
```

```python
#we do the predictions.
y_pred= model.predict(X_test)
```

```python
#comparison between the test predictions and actual test data.
y_test
```

```python
#there are significant deviations, but the values are somewhat in a ballpark.
```

```python
#######################################  Accuracy Metrics for Predicted Prices of Electric Vehicle    #############################
```

```python
#Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error
print("MAE: ",mean_absolute_error(y_test,y_pred))
```

```python
#Mean Squared Error(MSE)
from sklearn.metrics import mean_squared_error
print("MSE: ",mean_squared_error(y_test,y_pred))
```

```python
#Root Mean Squared Error(RMSE)
print("RMSE: ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#R Squared (R2)
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print("R2: ",r2)
```

```python
#In simple words:
#MAE → “Average error in plain units.”
#MSE → “Average squared error (big mistakes hurt more).”
#RMSE → “Average error in plain units, but penalizes big errors more.”
#R² → “How much of the variation is explained by the model.”

#Role of each metric
#MAE → tells you the average size of the error in the same units as the target.
#Example: “On average, my price prediction is off by 2000 pounds.”

#MSE → like MAE but punishes larger errors more strongly. Useful if big mistakes are especially bad in your problem.

#RMSE → makes MSE more interpretable (same units as target). Also highlights large errors.

#R² → tells you how much of the variation in the target data is explained by the model.
#Example: “My model explains 85% of the variation in car prices.”
```

```python
#drive type classification(random forest classification)

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Features and Target
X = df[['KWH','Acceleration','TopSpeed','Range','Efficiency','FastChargeSpeed','NumberofSeats','PriceinGermany']]
y = df['Drive']

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Train model
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)


```

```python
# Predictions
y_pred = model.predict(X_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#cm=confusion_matrix(y_test,y_pred,labels=[0,1])
print("Confision matrix\n",cm)


```

```python
plt.imshow(cm,cmap=plt.cm.Blues)
plt.colorbar()
plt.xlabel("Prediction value")
plt.ylabel("Actual value")
plt.show()

```
